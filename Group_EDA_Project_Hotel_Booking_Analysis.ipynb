### Variables Description

**The columns and the data it represents are listed below:**

**hotel** : Name of the hotel (Resort Hotel or City Hotel)

**is_canceled** : If the booking was canceled (1) or not (0)

**lead_time**: Number of days before the actual arrival of the guests

**arrival_date_year** : Year of arrival date

**arrival_date_month** : Month of month arrival date

**arrival_date_week_number** : Week number of year for arrival date

**arrival_date_day_of_month** : Day of arrival date

**stays_in_weekend_nights** : Number of weekend nights (Saturday or Sunday) spent at the hotel by the guests.

**stays_in_week_nights** : Number of weeknights (Monday to Friday) spent at the hotel by the guests.

**adults** : Number of adults among guests

**children** : Number of children among guests

**babies** : Number of babies among guests

**meal** : Type of meal booked

**country** : Country of guestsAnswer Here

**market_segment** : Designation of market segment

**distribution_channel** : Name of booking distribution channel

**is_repeated_guest** : If the booking was from a repeated guest (1) or not (0)

**previous_cancellations** : Number of previous bookings that were cancelled by the customer prior to the current booking

**previous_bookings_not_canceled** : Number of previous bookings not cancelled by the customer prior to the current booking

**reserved_room_type** : Code of room type reserved

**assigned_room_type** : Code of room type assigned

**booking_changes** : Number of changes/amendments made to the booking

**deposit_type** : Type of the deposit made by the guest

**agent** : ID of travel agent who made the booking

**company** : ID of the company that made the booking

**days_in_waiting_list** : Number of days the booking was in the waiting list

**customer_type** : Type of customer, assuming one of four categories

**adr** : Average Daily Rate, as defined by dividing the sum of all lodging transactions by the total number of staying nights

**required_car_parking_spaces** : Number of car parking spaces required by the customer

**total_of_special_requests** : Number of special requests made by the customer

**reservation_status** : Reservation status (Canceled, Check-Out or No-Show)

**reservation_status_date** : Date at which the last reservation status was updated

**total_night_stay** : Number of total nights (week night+ weekend night) spent at the hotel by the guests.

**revenue**: Revenue genrated by the customer calculated by adr * total_night_spent.

**total_guest** : Total no of Guest stayed in the hotel

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Finding NULL value Count
null_values= Hotel_Booking_Data.isnull().sum().sort_values(ascending=False).head()
null_values
````

````python
# Finding Null values Percentage in Company Column
percentage_of_null_values_company= null_values['company']/Hotel_Booking_Data.shape[0]*100
print(percentage_of_null_values_company)
````

````python
# Finding Null values Percentage in Agent Column
percentage_of_null_values_agent= null_values['agent']/Hotel_Booking_Data.shape[0]*100
print(percentage_of_null_values_agent)
````

````python
# As Agent column has 13% null values we'll fill those null value with value=0
Hotel_Booking_Data['agent'].fillna(value=0, inplace=True)
Hotel_Booking_Data['agent'].isnull().sum()
````

````python
# Finding Null values Percentage in Country Column
percentage_of_null_values_country= null_values['country']/Hotel_Booking_Data.shape[0]*100
print(percentage_of_null_values_country)
````

````python
# As country column has 0.51% null values we'll fill those null value with value='other'
Hotel_Booking_Data['country'].fillna(value='Other',inplace=True)
Hotel_Booking_Data['country'].isnull().sum()
````

````python
# As Agent column has very few % null values we'll fill those null value with value=0
Hotel_Booking_Data['children'].fillna(value=0,inplace=True)
Hotel_Booking_Data['children'].isnull().sum()
````

````python
# As company column has 93% null values we'll drop this column
Hotel_Booking_Data.drop(['company'],axis=1, inplace=True)
Hotel_Booking_Data.info()
````

````python
# Here we found the float type is of no use in column agent and children so we changed there date type
Hotel_Booking_Data[['children','agent']]= Hotel_Booking_Data[['children','agent']].astype('int64')
````

````python
# Added new columns for our visualization
Hotel_Booking_Data['total_night_stay']= Hotel_Booking_Data['stays_in_weekend_nights']+ Hotel_Booking_Data['stays_in_week_nights']
Hotel_Booking_Data['revenue']= Hotel_Booking_Data['adr']* Hotel_Booking_Data['total_night_stay']
Hotel_Booking_Data['total_guest']= Hotel_Booking_Data['adults']+Hotel_Booking_Data['children']+ Hotel_Booking_Data['babies']
Hotel_Booking_Data['reservation_status_date']= pd.to_datetime(Hotel_Booking_Data['reservation_status_date'])
Hotel_Booking_Data['year']= pd.DatetimeIndex(Hotel_Booking_Data['reservation_status_date']).year
Hotel_Booking_Data['month'] = pd.to_datetime(Hotel_Booking_Data['reservation_status_date'], format='%Y%m').dt.strftime('%B')
Hotel_Booking_Data['is_canceled']= Hotel_Booking_Data['is_canceled'].replace([0,1],['not canceled','canceled'])
Hotel_Booking_Data['is_repeated_guest']= Hotel_Booking_Data['is_repeated_guest'].replace([0,1],['Not a repeated guest','repeated guest'])
Hotel_Booking_Data
````

````python
Hotel_Booking_Data.info()
````

1) We deleted duplicated values from our Dataset

3) We wanted to changed the data type of column agent and children from float to Int and we      changed the data type of reservation_status_date colomn from object to datetime

4) we have replaced 0 and 1 with 'not canceled' and 'canceled' and 'Not a repeated guest' and 'repeated guest' in coloumn 'is_canceled' and 'is_repeated' respectively

#**Chart - 1. Per year revenue of each hotel**#

````python
# Chart - 1 visualization code

avg_rev = Hotel_Booking_Data.groupby(['year', 'hotel'])['revenue'].sum().reset_index()

plt.figure(figsize=(16, 9))
sns.set_style('whitegrid')
ax1 = sns.lineplot(data=avg_rev, x='year', y='revenue', hue='hotel')
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('Year', fontsize=16, style='italic')
plt.ylabel('Revenue', fontsize=16, style='italic')
plt.title('Year - Revenue Plot for Each Hotel', fontsize=18, style='oblique')
plt.xticks(style='oblique', fontsize=14)
plt.yticks(style='oblique', fontsize=14)

for line in ax1.lines:
    x_data, y_data = line.get_data()
    for x, y in zip(x_data, y_data):
        plt.text(x, y, f'{int(y):,}', ha='right', va='bottom', fontsize=11, fontweight='bold')
plt.show()
````
